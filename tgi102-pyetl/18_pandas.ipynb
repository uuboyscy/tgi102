```python
"""
data = [
    ['推: 0', '噓: 0', '分數: 0', '作者: Lineage097 (狐狸壽司)', ...],
    [],
    []
]
"""

import glob

data = list()
for path in glob.glob('./res_gossiping/*.txt'):
    try:
        tmpData = open(path, 'r', encoding='utf-8').read().split("---split---")[1].split("\n")
        data.append(tmpData[1:-1])
    except:
        pass
data
```

```python
columns = [c.split(": ")[0] for c in data[0]]
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(x):
    return x.split(": ")[1]
```
